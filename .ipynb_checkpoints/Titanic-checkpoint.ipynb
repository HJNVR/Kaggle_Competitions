```python
import numpy as np
import pandas as pd
```

```python
train_df = pd.read_csv("/Users/Starkjing/Desktop/Kaggle/titanic/train.csv")
test_df = pd.read_csv("/Users/Starkjing/Desktop/Kaggle/titanic/test.csv")
combine = [train_df, test_df] # combine is a list, index 0 is train, index 1 is test
train_df.head()
```

```python
train_df.isnull().sum(axis = 0)
```

```python
train_df.describe()
```

```python
test_df.isnull().sum(axis=0)
```

```python
for dataset in combine:    
    #complete missing age with median
    dataset['Age'].fillna(dataset['Age'].median(), inplace = True)

    #complete embarked with mode
    dataset['Embarked'].fillna(dataset['Embarked'].mode()[0], inplace = True)

    #complete missing fare with median
    dataset['Fare'].fillna(dataset['Fare'].median(), inplace = True)

drop_column = ['PassengerId','Cabin', 'Ticket', 'Name']
train_df.drop(drop_column, axis=1, inplace = True)

print(train_df.isnull().sum())
```

```python
test_df.drop(drop_column, axis=1, inplace=True)
test_df.isnull().sum()
#test_df.head()
```

```python
train = pd.get_dummies(train_df)
test = pd.get_dummies(test_df)
print(train.head())
print(test.head())
```

```python
from sklearn.ensemble import RandomForestClassifier

Y = train["Survived"]
X = train.iloc[:,1:]

model = RandomForestClassifier(n_estimators=100, max_depth=5, random_state=1)
model.fit(X,Y)

feature_imp = pd.Series(model.feature_importances_,index=list(X)).sort_values(ascending=False)
feature_imp # rank the important features
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
# Creating a bar plot
sns.barplot(x=feature_imp, y=feature_imp.index)
# Add labels to your graph
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title("Visualizing Important Features")
#plt.legend()
plt.show()
# remove sbSp , Parch, A
```
